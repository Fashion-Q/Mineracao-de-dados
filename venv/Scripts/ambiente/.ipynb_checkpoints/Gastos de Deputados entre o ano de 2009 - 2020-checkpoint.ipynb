```python
import pandas as pd
```

```python
df = pd.read_csv('gastos-deputados.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
#Remover as colunas que não irei utilizar
df = df.drop(columns=['datemissao'], errors='ignore')
df = df.drop(columns=['txtcnpjcpf'], errors='ignore')
```

```python
#Renomear o nome das colunas para ficar mais legível
df = df.rename(columns={'numano': 'Ano'}, errors='ignore')
df = df.rename(columns={'nummes': 'Mês'}, errors='ignore')
df = df.rename(columns={'txtdescricao': 'Descrição'}, errors='ignore')
df = df.rename(columns={'txtfornecedor': 'Fornecedor'}, errors='ignore')
df = df.rename(columns={'txnomeparlamentar': 'Nome'}, errors='ignore')
df = df.rename(columns={'sgpartido': 'Partido'}, errors='ignore')
df = df.rename(columns={'sguf': 'Estado'}, errors='ignore')
df = df.rename(columns={'vlrdocumento': 'Valor'}, errors='ignore')
df = df.rename(columns={'vlrglosa': 'Glosado'}, errors='ignore')
df = df.rename(columns={'vlrliquido': 'Total'}, errors='ignore')
```

```python
df.head()
```

```python
df[['Valor','Glosado', 'Total']].describe()
```

```python
df['Valor Comparavel'] = df['Valor'] - df['Glosado']
```

```python
#Considerar apenas 3 casas decimais porque alguns valores estavam
#aparecendo mesmo sendo iguais. Ex:160,33 era diferente de 160,334... tava estranho
df.loc[df['Total'].round(2) == df['Valor Comparavel'].round(2)]
```

```python
df.loc[df['Total'].round(2) != df['Valor Comparavel'].round(2)].sort_values(by='Valor Comparavel', ascending=False)
```

```python
df.loc[(df['Total'].round(2) != df['Valor Comparavel'].round(2)) &  (df['Nome'] == 'JONY MARCOS')].sort_values(by='Valor Comparavel', ascending=False)
```

```python
df.isna().sum()
```

```python
df.loc[df['Valor'].isna()]
```

```python
df.dropna(subset=['Valor'], inplace=True)
df.isna().sum()
```

```python
df['Nome'].nunique()
```

```python
df.loc[df['Valor'] < 0]
```

```python
df = df.drop('Valor', axis=1)
df = df.drop('Glosado', axis=1)
df = df.drop('Valor Comparavel', axis=1)
```

```python
df.head()
```

```python
df[df['Nome'] == 'JONY MARCOS'].groupby('Ano')['Total'].sum()
```

```python
gastos_por_partido_mes = df.groupby(['Ano', 'Mês', 'Partido','Descrição','Fornecedor'])['Total'].sum().reset_index()
gastos_por_partido_mes.rename(columns={'Total': 'Gasto Por Mês'}, inplace=True)
```

```python
gastos_por_partido_mes.sort_values(by='Gasto Por Mês', ascending=False)
```

```python
#Trabalhar com dados já rotulados Ex: Partido, Estado, Descrição...
bins = [-12000, 0 ,      15000, 35000 , float('inf')]
labels = ['Devolvido','Baixo', 'Médio', 'Muito Alto']

gastos_por_partido_mes['Nivel de Gastos'] = pd.cut(gastos_por_partido_mes['Gasto Por Mês'], bins=bins, labels=labels)
#pd.cut é uma função que categoriza os dados baseando-se nos intervalos. Cada elemento em 'Total' é colocado em
#um intervalo correspondente, e a função retorna uma Series categórica.

gastos_por_partido_mes.sort_values(by='Gasto Por Mês', ascending=False)
```

```python
gastos_por_pessoa_ano  = df.groupby(['Ano','Mês', 'Nome','Partido'])['Total'].sum().reset_index()
```

```python
gastos_por_pessoa_ano.sort_values(by='Total', ascending=False)
```

```python
gastos_por_partido_ano  = df.groupby(['Ano', 'Partido'])['Total'].sum().reset_index()
```

```python
gastos_por_partido_ano.sort_values(by='Total', ascending=False)
```

```python
gastos_por_partido_mes.sort_values(by='Gasto Por Mês', ascending=False)
```

```python
gastos_por_partido_mes = gastos_por_partido_mes.drop(['Descrição'], axis=1)
gastos_por_partido_mes = gastos_por_partido_mes.drop(['Fornecedor'], axis=1)
gastos_por_partido_mes[['Ano', 'Mês']] = gastos_por_partido_mes[['Ano', 'Mês']].astype(int)
```

```python
gastos_por_partido_mes.sort_values(by='Nivel de Gastos', ascending=False)
```

```python
#Transformar os dados categoricos em numericos, o pandas já tem
#biblioteca que faz isso que sao os dummies
partido_dummies = pd.get_dummies(gastos_por_partido_mes['Partido']).astype(int)
X = pd.concat([gastos_por_partido_mes['Ano'],gastos_por_partido_mes['Mês'],gastos_por_partido_mes['Gasto Por Mês'], partido_dummies], axis=1)
y = gastos_por_partido_mes['Nivel de Gastos']
```

```python
X
```

```python
y
```

```python
import plotly.graph_objects as go
import numpy as np

from sklearn import datasets
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

# É ou não é: Classificação
# Faz parte ou não faz parte... Clusterização
# Regressão é QUANDO
# Problema de Regressão -> Quanto de produto vai comprar
```

```python
#fazer dataframe de teste e treino
#tamanho do teste significa teste e treino: 0.3 = 30% de teste
#Size = 0.1 avaliar o modelo 10 vezes
#Fazendo uma previsão
x_treino, x_teste, y_treino, y_teste = train_test_split(X, y, test_size=0.1, random_state=42)
```

```python
y_teste
```

```python
x_treino
```

```python
y_treino
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
```

```python
knn.fit(x_treino, y_treino)
```

```python
resultado = knn.predict(x_teste)
resultado
```

```python
np.unique(resultado)
```

```python
x_teste
```

```python
#f1-score calcula precisao e recall junto, calcula a media harmonica entre eles
print(metrics.classification_report(y_teste,resultado, target_names=gastos_por_partido_mes['Nivel de Gastos'].unique()))
```
